# **a) Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” y la columna “Referencia”**

```python
# Carga de archivo con los "headers" deseados
df = pd.read_excel('Inmuebles.xlsx', skiprows = 1) # Saltar líneas
df = df.drop(['Referencia'], axis = 1) # 1 para columnas y 0 para las filas
df
```

```python
filtro1 = df.copy()
filtro1
```

```python
# Filtro a CSV
filtro1.to_csv("Filtro1.csv")
```

```python
df.info()
df.head()
```

# **b) Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00**

```python
# Convertir de fecha a "Object"
df['Fecha_Alta'] = df['Fecha_Alta'].astype(str)
```

```python
filtro2 = df[(df["Fecha_Alta"] == "2007-04-29") | (df["Fecha_Alta"] == "2004-01-02")]
filtro2
```

```python
# Filtro a CSV
filtro2.to_csv("Filtro2.csv")
```

# **c) Todos los registros de “ Tipo: parking, local, industrial y piso”**

```python
filtro3 = df[(df["Tipo"] == "Parking") | (df["Tipo"] == "Local")| (df["Tipo"] == "Industrial")| (df["Tipo"] == "Piso")]
filtro3
```

```python
# Filtro a CSV
filtro3.to_csv("Filtro3.csv")
```

# **d) Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”**

```python
filtro4 = df[(df["Provincia"] == "Girona") & (df["Superficie"] > 100)]
filtro4
```

```python
# Filtro a CSV
filtro4.to_csv("Filtro4.csv")
```

# **e) Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”**

```python
filtro5 = df[(df["Precio Venta"] > 1000000) | ((df["Vendedor"] == "Carmen") | (df["Vendedor"] == "Joaquin"))]
filtro5
```

```python
# Filtro a CSV
filtro5.to_csv("Filtro5.csv")
```

# **f) Todos los registros que no sean de “Provincia : Barcelona“**

```python
filtro6 = df[~(df["Provincia"] == "Barcelona")]
filtro6
```

```python
# Filtro a CSV
filtro6.to_csv("Filtro6.csv")
```

```python
filtro6['Provincia'].value_counts()
```

# **g)Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“**

```python
filtro7 = df[~(df["Operación"] == "Alquiler") & ~(df["Tipo"] == "Oficina")]
filtro7
```

```python
# Filtro a CSV
filtro7.to_csv("Filtro7.csv")
```

```python
filtro7['Operación'].value_counts()
filtro7['Tipo'].value_counts()
```
