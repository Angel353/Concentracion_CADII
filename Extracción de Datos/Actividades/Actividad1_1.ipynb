```python
# Se importa la librería
import pandas as pd
```

```python
# Cargar archivo csv o xlsx desde equipo
from google.colab import files
files.upload()
```

```python
# Carga desde archivo .csv o .xlsx sin índice
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

# **Filtro1 (Filas) y .csv**

```python
# Filtro por filas
filtro1 = df.iloc[0:6, :]
filtro1
```

```python
# Filtro a CSV
filtro1.to_csv("Filtro1.csv")
```

# **Filtro2 (Columnas) y .csv**

```python
# Filtro por columnas
filtro2 = df.iloc[:, [2,3,4]]
filtro2
```

```python
# Filtro a CSV
filtro2.to_csv("Filtro2.csv")
```

# **Filtro3,4,5 (Variables Cualitativas) y .csv**

```python
filtro3 = df[df['FECHA_DOC'] == '2022-01-05']
filtro3
```

```python
# Filtro a CSV
filtro3.to_csv("Filtro3.csv")
```

```python
filtro4 = df[df['NOMBRE_CLIENTE'] == 'PÚBLICO EN GENERAL']
filtro4
```

```python
# Filtro a CSV
filtro4.to_csv("Filtro4.csv")
```

```python
filtro5 = df[df["CVE_ART"] == 'TP0007']
filtro5
```

```python
# Filtro a CSV
filtro5.to_csv("Filtro5.csv")
```

# **Filtro6,7,8 (Variables Cuantitativas) y .csv**

```python
filtro6 = df[df["PRECIO_UNITARIO"] > 1000]
filtro6
```

```python
# Filtro a CSV
filtro6.to_csv("Filtro6.csv")
```

```python
filtro7 = df[df["COSTO_UNITARIO"] == 0]
filtro7
```

```python
# Filtro a CSV
filtro7.to_csv("Filtro7.csv")
```

```python
filtro8 = df[df["MARGEN_TOTAL_CALCULADO"] >= 20000]
filtro8
```

```python
# Filtro a CSV
filtro8.to_csv("Filtro8.csv")
```

# **Filtro9 (Conector Lógico) y .csv**

```python
# Filtro con 'Y'

filtro9 = df[(df["PRECIO_UNITARIO"] > 3000) & (df["COSTO_UNITARIO"] <= 1000) & (df["COSTO_UNITARIO"] > 0)]
filtro9
```

```python
# Filtro a CSV
filtro9.to_csv("Filtro9.csv")
```
