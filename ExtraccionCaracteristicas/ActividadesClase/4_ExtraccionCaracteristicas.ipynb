## **Modelado explicativo de variables numéricas**

```python
# Ajustar máximo de filas
pd.options.display.max_rows = None
```

```python
# Corroboramos valores nulos
valoresNulos = Micro_Retailer.isnull().sum()
valoresNulos
```

```python
#Calculamos el número total de la población "n"

Micro_Retailer['_latitude'].info()
n = len(Micro_Retailer['_latitude'])
```

```python
# Obtener el límite superior y el límite inferior de la columna objetivo
Max = Micro_Retailer["_latitude"].max()
Min = Micro_Retailer["_latitude"].min()
Limites = [Min, Max]
Limites
```

```python
# Calculamos el rango R
R = Max - Min
R
```

```python
# Regla de Sturges: Calcular el número de Intervalos de Clase "ni"
ni = 1 + 3.32*np.log10(n)
ni
```

```python
# Calculamos el ancho del intervalo "i"
i = R/ni
i
```

## **Creación de intervalos a partir de clases**

```python
# Categorización de variables
intervalos = np.linspace(Min, Max, 9)
intervalos
```

```python
# Creamos las categorías
categorias = ["Categoría1", "Categoría2", "Categoría3", "Categoría4","Categoría5", "Categoría6", "Categoría7", "Categoría8"]
```

```python
# Finalmente creamos las categorías en la columna numérica
Micro_Retailer['_latitude'] = pd.cut(x = Micro_Retailer['_latitude'], bins = intervalos, labels = categorias)
Micro_Retailer['_latitude']
```
