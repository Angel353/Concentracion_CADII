# **Filtros básicos para variables cualitativas**

```python
# Filtro por objeto

filtro1 = df[df['FECHA_VENCIMIENTO'] == '2023-04-13']
filtro1
```

# **Filtros por filas y columnas iloc**

```python
# Filtro por filas
filtro2 = df.iloc[3:10, :] # Filas consecutivas (De la fila 3 a la 10 y todas las columnas)
filtro2 = df.iloc[[8, 24, 100], :] # Filas no consecutivas (Filas primero y luego columnas)
filtro2
```

```python
# Filtro por columnas
# filtro3  = df.iloc[:, 1:3] # Columnas consecutivas
filtro3  = df.iloc[10:20, [1,4,5]] # Columnas no consecutivas
filtro3
```

# **Colocar columna como índice**

```python
# Colocar columna como índice
df1 = pd.read_excel('Antigüedad de saldos 17.03.2023.xlsx', index_col = 1)
df1
```

```python
# Paso 2: Aplicamos filtro de filas vs columnas
filtro4 = df1.loc[["HELVEX", "DONGA HWASUNG MEXICO"], ["MONTO ADEUDADO"]] # Con el índice en "nombre"
filtro4
```

```python
# Filtro por cabecera y al final de los datos
# filtro5 = df.head(10)
filtro5 = df.tail()
filtro5
```

# **Filtro para variables cuantitativas**

```python
# Filtro por comparación
filtro6 = df[df["MONTO ADEUDADO"] > 70000]
filtro6
```

# **Filtro por conectores lógicos**

```python
# Filtro con 'Y'

filtro7 = df[(df["MONTO ADEUDADO"] < 6000) & (df["FECHA_FACTURA"] == "2023-03-14")]
filtro7
```

```python
# Filtro con 'o'

filtro8 = df[(df["MONTO ADEUDADO"] < 6000) | (df["FECHA_FACTURA"] == "2023-03-14")]
filtro8
```

```python
# Filtro not "~"

filtro9 = df[~(df["MONTO ADEUDADO"] < 6000) & ~(df["FECHA_FACTURA"] == "2023-03-14")]
filtro9
```

```python
# Convertir archivo filtrado a CSV
filtro9.to_csv("Entregable1.csv")
```
